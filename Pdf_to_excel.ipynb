```python
import re
import parse
import pdfplumber
import pandas as pd
from collections import namedtuple
```

```python
Line = namedtuple('Line', 'model rated_capacity length_of_would_depth maximum_operation_pressure overall_height outside_diameter f2f_dimension pipe_size shiping_weight')
```

```python
line_re = re.compile(r'\d\d\d \d\d \d\d\d')
```

```python
file='1 FULFLO SINGLE.pdf'
```

```python
lines = []

with pdfplumber.open(file) as pdf:
    pages = pdf.pages
    
    for page in pdf.pages:
        text = page.extract_text(x_tolerance=2,y_tolerance=0)
        for line in text.split('\n'):
            print(line)
            if line_re.search(line):
                print(line)
                model = line.split()[:2]
                model= ' '.join(model)
                rated_capacity = line.split()[2:3]
                rated_capacity= ''.join(rated_capacity)
                length_of_would_depth = line.split()[3:4]
                length_of_would_depth= ''.join(length_of_would_depth)
                maximum_operation_pressure = line.split()[4:5]
                maximum_operation_pressure= ''.join(maximum_operation_pressure)
                maximum_operation_pressure= maximum_operation_pressure.replace('(','')
                overall_height = line.split()[6:7]
                overall_height= ''.join(overall_height)
                outside_diameter = line.split()[7:8]
                outside_diameter= ''.join(outside_diameter)
                f2f_dimension = line.split()[8:9]
                f2f_dimension= ''.join(f2f_dimension)
                pipe_size = line.split()[9:10]
                pipe_size= ''.join(pipe_size)
                shiping_weight = line.split()[10:11]
                shiping_weight= ''.join(shiping_weight)
                info_line=Line(model, rated_capacity, length_of_would_depth, maximum_operation_pressure, overall_height, outside_diameter, f2f_dimension, pipe_size, shiping_weight)
                lines.append(info_line)
```

```python
df = pd.DataFrame(lines)
df.head(2)
```
